# 1. Loi continues
## 1.1 Loi uniforme
L'une des loi les plus simples et utiles est peut-être la distribution uniforme. La fonction de densité la loi uniforme continue est:
$$
f(x) =
\begin{cases}
\frac{1}{b-a} \,\,\,\, \text{lorsque} x \in [a,b]\\
0 \,\,\,\, \text{sinon}\\
\end{cases} \ 
$$

Étant donné que tout intervalle de nombres d'égale largeur a une probabilité égale d'être observé, la courbe décrivant la distribution est un rectangle, avec une hauteur constante à travers l'intervalle et une hauteur nulle ailleurs. Étant donné que l'aire sous la courbe doit être égale à 1, la longueur de l'intervalle détermine la hauteur de la courbe. La figure suivante montre une distribution uniforme dans l'intervalle $(a, b)$. Notez que la zone doit être 1. La hauteur est définie sur $\frac{1}{b − a}$.

Vous pouvez visualiser une distribution uniforme en python à l'aide d'un générateur de nombres aléatoires agissant sur un intervalle de nombres $(a, b)$. Vous devez importer la fonction uniforme du module `scipy.stats`

La fonction uniforme génère une variable continue uniforme entre l'intervalle spécifié via ses arguments `loc` et `scale`. Cette distribution est constante entre loc et scale. Les arguments de `size` décrivent le nombre de variables aléatoires. Si vous souhaitez conserver la reproductibilité, incluez un argument random_state affecté à un nombre.

Vous pouvez utiliser la fonction `distplot` de Seaborn pour tracer l'histogramme de la distribution que vous venez de créer. Le `distplot` de Seaborn accepte plusieurs arguments pour personnaliser la figure. Vous créez d'abord un objet figure `ax`. Ici, vous pouvez spécifier le nombre de cases dans l'histogramme, spécifier la couleur de l'histogramme et spécifier l'option de tracé de densité avec `kde` et l'option de largeur de ligne avec `hist_kws`. Vous pouvez également définir des étiquettes pour les axes x et y à l'aide des arguments xlabel et ylabel.

```
# import uniform distribution
from scipy.stats import uniform

# random numbers from uniform distribution
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)

ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Loi uniforme ', ylabel='Fréquence')
```

### Note
Cette loi apparaît fréquemment sous sa forme discrète par exemple le lancé de dés. Elle apparaît, dès que les événements sont équiprobables, dans sa forme continue, c'est par exemple le cas lorsqu'on tire un nombre au hasard dans un intervalle.

## 1.2 Loi normale
La loi normale, également connue sous le nom de loi Gaussienne, est omniprésente dans la science des données. Vous le rencontrerez à de nombreux endroits, en particulier dans les sujets d'inférence statistique. C'est également l'une des hypothèses de nombreux algorithmes de science des données.

Une distribution normale a une courbe de densité en forme de cloche décrite par sa moyenne $\mu$ et son écart type $\sigma$. La courbe de densité est symétrique, centrée sur sa moyenne, son étalement étant déterminé par son écart type montrant que les données proches de la moyenne sont plus fréquentes que les données éloignées de la moyenne. La fonction de distribution de probabilité d'une courbe de densité normale avec la moyenne $\mu$ et l'écart type $\sigma$ en un point x donné est donnée par:

$$
f(x;\mu, \sigma^{2}) = \frac{e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}}{\sigma\sqrt{2\pi}}
$$



Près de 68% des données se situent à une distance d'un écart-type de la moyenne de chaque côté et 95% à deux écarts-types. Il convient également de mentionner qu'une loi avec une moyenne de 0 et un écart-type 1 est appelée une loi normale standard.

Vous pouvez générer une variable aléatoire normalement distribuée à l'aide de la méthode `norm.rvs()` du module `scipy.stats`. L'argument `loc` correspond à la moyenne de la loi et `scale` correspond à l'écart type et `size` au nombre de variables aléatoires. Si vous souhaitez conserver la reproductibilité, incluez un argument `random_state` affecté à un nombre.

Vous pouvez visualiser la loi comme vous l'avez fait avec la distribution uniforme, en utilisant les fonctions de `distplot` de seaborn. La signification des arguments reste la même que dans le dernier cas.

```
from scipy.stats import norm
# generate random numbers from N(0,1)
data_normal = norm.rvs(size=10000,loc=0,scale=1)

ax = sns.distplot(data_normal,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 20,'alpha':1})
ax.set(xlabel='Loi normale', ylabel='Fréquence')
```

## 1.3 Loi gamma

La loi gamma est une famille à deux paramètres de loi de probabilités continues comme la loi normale. Bien qu'elle soit rarement utilisé dans sa forme brute, mais d'autres lois couramment utilisées comme les distributions exponentielles, khi 2, erlang sont des cas particuliers de la loi gamma. La distribution gamma peut être paramétrée en termes d'un paramètre de forme $\alpha = k$ et d'un paramètre d'échelle inverse $\beta = \frac{1}{\theta}$., Le symbole $\Gamma(n)$ est la fonction gamma et est défini comme $(n-1)!$ dans le cas discret.

$$
f(x;\alpha, \beta) = \frac{\beta^{\alpha}x^{\alpha-1}e^{-\beta x}}{\Gamma(\alpha)}
$$

Vous pouvez générer une variable aléatoire distribuée gamma en utilisant la méthode `gamma.rvs()` du module `scipy.stats` qui prend le paramètre de forme `a` comme argument. Quand `a` est un entier, gamma se réduit à la loi d'Erlang, et quand `a` = 1 à la distribution exponentielle. Pour déplacer la loi, utilisez l'argument `loc`, pour changer l'échelle, utilisez l'argument `scale`. Ici `scale` est identique à $\frac{1}{\beta}$. Le paramètre `size` détermine le nombre de variables aléatoires dans la distribution. Si vous souhaitez conserver la reproductibilité, incluez un argument `random_state` affecté à un nombre.

```
from scipy.stats import gamma
data_gamma = gamma.rvs(a=9, scale=.5, size=10000)

ax = sns.distplot(data_gamma,
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Loi gamma', ylabel='Fréquence')
```

## 1.4 Loi Exponentielle
La distribution exponentielle décrit le temps entre les événements dans un processus ponctuel de Poisson, c'est-à-dire un processus dans lequel les événements se produisent de manière continue et indépendante à un taux moyen constant. Il a un paramètre $\lambda$ appelé paramètre de débit, et son équation est décrite comme:

$$
f(x) =
\begin{cases}
\lambda e^{-\lambda x} \,\,\,\, \text{lorsque} \,\,\,\, x \geq 0\\
0 \,\,\,\, \text{sinon}\\
\end{cases} \ 
$$

Vous pouvez générer une variable aléatoire distribuée suivant une loi exponentielle en utilisant la méthode `expon.rvs()` du module `scipy.stats` qui prend le paramètre `scale` qui n'est rien d'autre que $\frac{1}{\lambda}$ dans l'équation. Pour déplacer la loi, utilisez l'argument `loc`, `size` décide du nombre de variables aléatoires dans la distribution. Si vous souhaitez conserver la reproductibilité, incluez un argument `random_state` affecté à un nombre.

```
from scipy.stats import expon
data_expon = expon.rvs(scale=1,loc=0,size=1000)

ax = sns.distplot(data_expon,
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Loi exponentielle', ylabel='Fréquence')
```

# 2. Loi discrètes
## 2.1 Loi de Poisson
La variable aléatoire de Poisson est généralement utilisée pour modéliser le nombre de fois qu'un événement s'est produit dans un intervalle de temps. Par exemple, on peut modéliser le nombre d'utilisateurs visitant un site Web dans un intervalle de 10 minutes avec une loi de Poisson.  Un événement peut se produire $0, 1, 2,\cdots, n$ fois dans un intervalle. Le nombre moyen d'événements dans un intervalle est désigné par $\lambda$ (lambda). La probabilité d'observer k événements dans un intervalle est donnée par l'équation:

$$
\mathbb{P}(X=k;\lambda) = e^{-\lambda}\frac{\lambda^{-k}}{k!}
$$

Notez que la loi normale est un cas limite de loi de Poisson avec le paramètre $\lambda \rightarrow \infty$. De plus, si les temps entre les événements aléatoires suivent une loi exponentielle avec le taux $\lambda$, alors le nombre total d'événements dans une période de temps t suit la distribution de Poisson avec le paramètre $\lambda t$.

```
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=5.5, size=10000)

ax = sns.distplot(data_poisson,
                  bins=30,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Loi de Poisson', ylabel='Fréquence')
```

## 2.2 Loi de Bernoulli
Une loi de Bernoulli n'a que deux résultats possibles, à savoir 1 (succès) et 0 (échec), et un seul essai, par exemple, un tirage au sort. Ainsi, la variable aléatoire $X$ qui a une loi de Bernoulli peut prendre la valeur 1 avec la probabilité de succès, p, et la valeur 0 avec la probabilité d'échec, q ou 1 − p. Les probabilités de réussite et d'échec ne sont pas nécessairement aussi probables. La loi de Bernoulli est un cas particulier de la loi binomiale où un seul essai est réalisé (n = 1). Sa loi de probabilité est donnée par:

$$
\mathbb{P}(X=k;p) = p^{k}(1-p)^{1-k} \,\,\,\,\text{avec} \,\,\,\, k\in\{0,1\}
$$

Vous pouvez générer une variable aléatoire discrète distribuée suivant une loi de Bernoulli en utilisant la méthode `bernoulli.rvs()` du module `scipy.stats` qui prend $p$ (probabilité de succès) comme paramètre. Pour décaler la loi, utilisez le paramètre `loc`. la taille décide du nombre de répétitions des essais. Si vous souhaitez conserver la reproductibilité, incluez un argument `random_state` affecté à un nombre.

```
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6)

ax= sns.distplot(data_bern,
                 kde=False,
                 color="skyblue",
                 hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Loi de Bernoulli', ylabel='Fréquence')
```

## 2.3 Loi binomiale
Une loi où seuls deux résultats sont possibles, comme le succès ou l'échec, le gain ou la perte, et où la probabilité de succès et d'échec est la même pour tous les essais est appelée une loi binomiale. Cependant, les résultats ne sont pas nécessairement aussi probables et chaque essai est indépendant des autres. Les paramètres d'une loi binomiale sont $n$ et $p$ où $n$ est le nombre total d'essais et $p$ est la probabilité de succès de chaque essai. Sa loi de probabilité est donnée par:

$$
\mathbb{P}(X=k;n,p) = \binom nk p^{k}(1-p)^{n-k} \,\,\,\, \text{avec} \,\,\,\, \binom nk = \frac{n!}{(n-k)!k!}
$$

Vous pouvez générer une variable aléatoire discrète distribuée binomiale en utilisant la méthode `binom.rvs()` du module `scipy.stats` qui prend $n$ (nombre d'essais) et $p$ (probabilité de succès) comme paramètres. Pour décaler la distribution, utilisez le paramètre `loc`. la taille décide du nombre de répétitions des essais. Si vous souhaitez conserver la reproductibilité, incluez un argument `random_state` affecté à un nombre.

```
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)

ax = sns.distplot(data_binom,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Loi binomiale', ylabel='Fréquence')
```

Notez que la probabilité de réussite étant supérieure à 0.5, la loi est biaisée vers le côté droit. Par ailleurs, la loi de Poisson est un cas limite d'une loi binomiale dans les conditions suivantes:

* Le nombre d'essais est indéfiniment grand ou lorsque $n \rightarrow \infty$.
* La probabilité de succès de chaque essai est identique et infiniment petite ou $p \rightarrow 0$.
* np = λ, est fini.

La loi normale est une autre forme limitative de la loi binomiale dans les conditions suivantes:

* Le nombre d'essais est indéfiniment grand, i.e $n \rightarrow \infty$.
* $p$ n'est pas infiniment petit.
